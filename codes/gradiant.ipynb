```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(X, y, w, b):
    n = len(y)
    y_pred = np.dot(X, w) + b
    cost = (1/(2*n)) * np.sum((y_pred - y)**2)
    return cost

# Generate some random data
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.random.randint(0, 2, size=(100,))

# Define the range of weights and bias to plot
w_range = np.linspace(-2, 2, num=100)
b_range = np.linspace(-2, 2, num=100)

# Create a grid of weights and bias
w_grid, b_grid = np.meshgrid(w_range, b_range)

# Calculate the cost function for each combination of weights and bias
cost_grid = np.empty_like(w_grid)
for i in range(len(w_range)):
    for j in range(len(b_range)):
        w = np.array([w_range[i], w_range[j]])
        b = b_range[j]
        cost = cost_function(X, y, w, b)
        cost_grid[j, i] = cost

# Plot the cost function as a contour plot
plt.contour(w_grid, b_grid, cost_grid, levels=20)
plt.scatter(w, w, marker='x', color='red')
plt.xlabel('Weight 1')
plt.ylabel('Weight 2')
plt.title('Cost Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(X, y, w, b):
    n = len(y)
    y_pred = np.dot(X, w) + b
    cost = (1/(2*n)) * np.sum((y_pred - y)**2)
    return cost

# Generate some random data
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.random.randint(0, 2, size=(100,))

# Define the range of weights and bias to plot
w_range = np.linspace(-2, 2, num=100)
b_range = np.linspace(-2, 2, num=100)

# Create a grid of weights and bias
w_grid, b_grid = np.meshgrid(w_range, b_range)

# Calculate the cost function for each combination of weights and bias
cost_grid = np.empty_like(w_grid)
for i in range(len(w_range)):
    for j in range(len(b_range)):
        w = np.array([w_range[i], w_range[j]])
        b = b_range[j]
        cost = cost_function(X, y, w, b)
        cost_grid[j, i] = cost

# Plot the cost function as a contour plot
plt.contour(w_grid, b_grid, cost_grid, levels=20)
plt.xlabel('Weight 1')
plt.ylabel('Weight 2')
plt.title('Cost Function')

# Find the indices of the minimum cost in the cost_list
min_cost_idx = np.argmin(cost_grid)

# Get the weights and bias corresponding to the minimum cost
w_min = np.array([w_grid.flatten()[min_cost_idx], 0])
b_min = b_grid.flatten()[min_cost_idx]

# Print the minimum cost and the corresponding weights and bias
print(f"Minimum cost: {cost_grid.flatten()[min_cost_idx]:.2f}")
print(f"Optimal weights: {w_min}")
print(f"Optimal bias: {b_min}")

# Plot the optimal weights and bias
plt.scatter(w_min[0], w_min[1], marker='x', color='red')

plt.show()
```

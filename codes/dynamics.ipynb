```python
import matplotlib.pylab as plt
from scipy.integrate import odeint
import numpy as np
```

```python
def dAdt(A, t, beta, gamma, N):
  S = A[0]
  I = A[1]
  R = A[2]
  return[-beta/N * S * I,
         beta/N * S * I - gamma * I,
         gamma * I]
```

```python
times = np.arange(0, 100)
gamma = 1/10
N = 10000000
beta = 0.39
S0 = N - 574
I0 = 574
R0 = 0
sol = odeint(dAdt, y0=[S0, I0, R0], t=times, args=(beta, gamma, N))

S = sol.T[0]
I = sol.T[1]
R = sol.T[2]

plt.plot(times, S,label ='S')
plt.plot(times, I,label ='I')
plt.plot(times, R,label = 'R')
plt.grid()
plt.legend()
plt.xlabel('T')
plt.ylabel('N')
```

```python
N = 1000000
S = N - 1
I = 1
beta = 0.3
gamma = 0.1
t = 100

sus = []
inf = []

def infection(S, I, N, t):
    for t in range (0, t):
        S = S - (beta*S*I/N) + gamma * I
        I = I + (beta*S*I/N) - gamma * I
        sus.append(S)
        inf.append(I)

infection(S, I, N, t)
plt.plot(sus, label='S')
plt.plot(inf, label='I')
plt.legend()
plt.grid()
plt.xlabel('T')
plt.ylabel('N')
```
